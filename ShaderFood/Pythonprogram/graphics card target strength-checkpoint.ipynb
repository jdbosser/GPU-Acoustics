## summation example ##

Imagine a sphere wit radius 2 meter.
View this through a gitter of points separated distance *d*

For each gitter point there is an associated area $d²$




```python
from math import *
import numpy as np
import time
import matplotlib.pyplot as plt


# Input parameters

R = 6.0         # radius
f = 2000.0     # frequency
c = 1500.0      # speed of sound

lamda = c/f     # wavelength
k = 2*pi/lamda  # wave number

# other parameters
mmax = 500
n = 2*mmax + 1    # array dimension
step = R/mmax     # steg mellan enskilda punkter (i x eller y)
pixarea = R*R/(n*n)

xs = np.linspace(-R,R,n)
ys = np.linspace(-R,R,n)
elems = np.zeros((n,n), dtype = np.complex128)      # element contributions

# z-koordinaten för punkt (i,j) 
# här tänker jag mig en indexering från -nmax till nmax
def z(R,x,y):
    try:
        sq = sqrt(R*R - x*x - y*y)  
    except:
        print('exception',  R, x ,y)
        sq = 0
    return sq

        
# normalvinkeln för punkt (i,j)
# här tänker jag mig en indexering från -nmax till nmax
def cosfi(R,x,y):
    return z(R,x,y)/R

#projicerad area för en pixel    
# här tänker jag mig en indexering från -nmax till nmax
def projarea(R,x,y):
    return pixarea/cosfi(R,x,y)

def inrange(R,x,y):
    return (x*x + y*y) < (R*R)


def fill_array():
    for i in range(n):
        x = xs[i]
        for j in range(n):
            y = ys[j]
            if inrange(R,x,y):
                try:
#                    a = projarea(x,y)
#                    a = pixarea*cosfi(x,y)
                    a = pixarea
                    val = 4*a/lamda  # intensiteten (varför 4*?)
                    phase = -2*k*(9-z(R,x,y))
                    elems[i,j] = val*complex(cos(phase), sin(phase))
                except:
                    print(x,y,a)


print('lambda', lamda)
start_time = time.time()
fill_array();    
fill_time = time.time()    
res = abs(np.sum(elems))**2
res_time = time.time()
print(res)
print('fill time', (fill_time - start_time))
print('sum time', (res_time - fill_time))

    
```

```python
def square_TS(a,b,n,frequency,angle = 0.0):
    N = 2*trunc(n/2) + 1    # I want an odd number
    xs = np.linspace(-a/2,a/2,N)
    ys = np.linspace(-b/2,b/2,N)
    lamda = c/frequency
    k = 2*pi/lamda
    elems = np.zeros((N,N), dtype = np.complex64)      # element contributions
    pixarea = a*b/(N*N)
    """
    print 'a',a
    print 'b',b
    print 'lambda', lamda
    print 'pixarea',pixarea
    """
    print(angle*180/pi)
    
    for i in range(N):
        x = xs[i]
        for j in range(N):
            y = ys[j]
            try:
#                a = projarea(x,y)
                a = pixarea
                val = a*cos(angle)/lamda  # intensiteten (varför inte *4 här)
                phase = -2*k*(x*sin(angle))
                elems[i,j] = val*complex(cos(phase), sin(phase))
            except:
                print(x,y,a)
    return abs(np.sum(elems))**2

c = 1500.0
a = 2.0
b = 1.02
freq = 10000.0
angle = 0
n = 1000
print(square_TS(a,b,n,freq, 0.0))

# ovanstående stämmer bra för vinkeln 0


angles = np.linspace(0,pi/2,900)
plt.plot(angles*180/pi, [10*log10(square_TS(a,b,n,freq,angle)) for angle in angles])
plt.show()

```
